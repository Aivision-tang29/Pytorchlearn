```python
import torch
from torch.autograd import Variable
from torch import nn
```

```python
rnn_single=nn.RNNCell(input_size=100,hidden_size=200)
```

```python
rnn_single.weight_hh.shape
```

```python
rnn_single.weight_ih.shape
```

```python
x=Variable(torch.randn(6,5,100))
```

```python
h_t=Variable(torch.zeros(5,200))
```

```python
out=[]
for i in range(6):
    h_t=rnn_single(x[i],h_t)
    out.append(h_t)
```

```python
h_t
```

```python
len(out)
```

```python
out[0].shape
```

```python
rnn_seq=nn.RNN(100,200)
```

```python
rnn_seq.weight_hh_l0.shape
```

```python
out,h_t=rnn_seq(x)
```

```python
h_t.shape
```

```python
lstm_seq.weight_hh_l0.shape
```

```python
lstm_seq.weight_ih_l0.shape
```

```python
lstm_seq.weight_hh_l1.shape
```

```python
lstm_seq.weight_ih_l1.shape
```

```python
lstm_input = Variable(torch.randn(10, 3, 50)) # 序列 10，batch 是 3，输入维度 50
```

```python
out, (h, c) = lstm_seq(lstm_input)
```
